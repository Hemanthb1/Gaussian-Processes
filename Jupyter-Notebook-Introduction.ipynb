## Speeding up your code -- Cython

Another way to optimise your code is to "turn in to C" with cython. It is a very usefull tool if you like coding in Python, but you like the speed of C, although it will never be as fast as a code written purely in C.



```python
print("Hello World")```

Save this code in a file named helloworld.pyx. Now we need to create the setup.py, which is like a python Makefile (for more information see Source Files and Compilation). Your setup.py should look like:
```python
from distutils.core import setup
from Cython.Build import cythonize

setup(
    ext_modules = cythonize("helloworld.pyx")
)```

To use this to build your Cython file use the commandline options:

$ python setup.py build_ext --inplace

Which will leave a file in your local directory called helloworld.so in unix or helloworld.pyd in Windows. Now to use this file: start the python interpreter and simply import it as if it was a regular python module:

```python
import helloworld
>>>Hello World```


Of course, there is a lot more you can do with Cython!

# Working on an Environment

If you want to try other Python configurations or different versions you can create new environments where you can install different Python versions from you main one, so it doesn't create any conflict with your existing version. You can create an environment and install any package you want.

    conda create --name myenv 
    
or    
    
    conda create -n myenv python=3.4
    conda install -n myenv scipy
    
The first line just creates the environment called _myenv_, the second line create the environment _myenv_ with Python 3.4 in it and the last name install scipy in the _myenv_ environment. You can also install packages inside an evironment by activating it and "normally" installing it with _pip_ or _conda_ inside it.

To activate an environment you need to run the next command in your terminal:

    source activate myenv
    
where _myenv_ is the name of the environment. To exit an environment just type:

    source deactivate
